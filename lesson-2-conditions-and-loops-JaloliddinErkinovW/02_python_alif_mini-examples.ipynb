```python
saved_pwd = "right_password"
pwd = input("Введите пароль для входа: ")
while pwd != saved_pwd:
    pwd = input("Введите пароль для входа: ")
print("Пароль верный. Вход разрешён.")
```

```python
saved_pwd = "right_password"
while input("Введите пароль для входа: ") != saved_pwd:
    pass
print("Пароль верный. Вход разрешён.")
```

```python
name = input("Введите имя: ")
while name != "СТОП":
    print(f"Привет, {name}!")
    name = input("Введите имя: ")
print("Программа завершена.")
```

```python
while (name := input("Введите имя: ")) != "СТОП":
    print(f"Привет, {name}!")
print("Программа завершена.")
```

```python
from math import asin, cos, radians, sin, sqrt

# Approximate radius of Earth in kilometers
rad = 6371

# Locations of Oslo and Vancouver
ϕ1, λ1 = radians(59.9), radians(10.8)
ϕ2, λ2 = radians(49.3), radians(-123.1)

# Distance between Oslo and Vancouver
2 * rad * asin(
    sqrt(
        sin((ϕ2 - ϕ1) / 2) ** 2
        + cos(ϕ1) * cos(ϕ2) * sin((λ2 - λ1) / 2) ** 2
    )
)
```

```python
# вложенные циклы
for i in range(26):
    for j in range(26):
        print(f"{chr(ord('a') + i)}{chr(ord('a') + j)}")# вложенные циклы
```

```python
password = "right_password"
while True:
    if input("Введите пароль: ") == password:
        print("Пароль принимается")
        break
```

```python
flag = False
for i in range(26):
    for j in range(26):
        text = f"{chr(ord('a') + i)}{chr(ord('a') + j)}"
        if text == "ya":
            print(text)
            flag = True
            continue
        print(text)
    if flag:
        continue
```

```python
for i in range(26):
    for j in range(26):
        if i == j:
            continue
        print(f"{chr(ord('a') + i)}{chr(ord('a') + j)}")
```

```python
for i in range(26):
    for j in range(26):
        if i != j:
            print(f"{chr(ord('a') + i)}{chr(ord('a') + j)}")
```

```python
# while + pass + else
while input("Введите строку (СТОП для остановки): ") != "СТОП":
    pass
else:
    print("Цикл завершён")
```

```python
# while + break + else
while (text := input("Введите строку (СТОП для остановки): ")) != "СТОП":
    if text == "ignore_else":
        break
else:
    print("Цикл завершён")
```
